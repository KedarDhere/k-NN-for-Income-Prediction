### Fit k-NN via Scikit Learn

```python
import pandas as pd
import numpy as np
import time
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from datetime import datetime as dt
```

```python
# Import the Training and Dev dataset
train_complete_data = pd.read_csv("income.train.txt.5k",sep=",", names=["age","sector","edu","marriage","occupation","race","sex","hours","country","target"])
dev_complete_data = pd.read_csv("income.dev.txt",sep=",", names=["age","sector","edu","marriage","occupation","race","sex","hours","country","target"])
```

```python
# Drop the target columns and separate data and labels
train_data = train_complete_data.drop(columns=["target"])
train_labels = train_complete_data["target"]

dev_data = dev_complete_data.drop(columns=["target"])
dev_labels = dev_complete_data["target"]
```

```python
# Perform Binazrization
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoder.fit(train_data)

train_binary_data = encoder.transform(train_data)
dev_binary_data = encoder.transform(dev_data)
```

```python
# Calculate the Positive Rate
def calculate_positive_rate(predicted_positive_rate):
    result = 0
    for i in range(len(predicted_positive_rate)):
        if " >50K" == predicted_positive_rate[i]:
            result += 1
    return result
```

```python
# Use sklearn KNeighborsClassifier to implement K-NN
max_dev_time = float('-inf')
min_dev_time = float('inf')

max_train_time = float('-inf')
min_train_time = float('inf')
for i in range(1, 100, 2):
    knn_model = KNeighborsClassifier(n_neighbors=i)
    knn_model.fit(train_binary_data, train_labels)
    
    start_time = time.time()
    prediction_train = knn_model.predict(train_binary_data)
    train_positive_rate = calculate_positive_rate(prediction_train)
    train_runningTime = (time.time() - start_time)
    accuracy_score_train = knn_model.score(train_binary_data, train_labels) * 100
    train_error = round(100 - accuracy_score_train,2) 
    
    start_time = time.time()
    prediction_dev = knn_model.predict(dev_binary_data)
    dev_positive_rate = calculate_positive_rate(prediction_dev)
    dev_runningTime = (time.time() - start_time)
    accuracy_score_dev = knn_model.score(dev_binary_data, dev_labels) * 100
    dev_error = round(100 - accuracy_score_dev,2)
    
    max_train_time = max(max_train_time, train_runningTime)
    min_train_time = min(min_train_time, train_runningTime)
    
    max_dev_time = max(max_dev_time, dev_runningTime)
    min_dev_time = min(min_dev_time, dev_runningTime)
    print(f'k = {i},  train_data_error={train_error}%,  dev_data_error = {dev_error}%, train_positive_rate={train_positive_rate}, dev_positive_rate={dev_positive_rate}, training_running_time={train_runningTime}, dev_running_time={dev_runningTime} ')
#     print(f'k = {i},  train_data_error={train_error}%,  dev_data_error = {dev_error}% ')
#     print(f'k = {i},  train_positive_rate={train_positive_rate}, dev_positive_rate={dev_positive_rate} ')

print(f'Max Training Running Time: {max_train_time}')
print(f'Min Training Running Time: {min_train_time}')
print(f'Max Dev Running Time: {max_dev_time}')
print(f'Min Training Running Time: {min_dev_time}')
```

### Implement your own k-NN classifier

```python
class KnnClassifier():
    def __init__(self, k=1, dist_cal_method = "eulidean"):
        self.k = k
        self.dist_cal_method = dist_cal_method
    
    def fit(self, xlabel, ylabel):
        self.X_train = xlabel
        self.Y_train = ylabel
        
    def calculate_euclidean_distance(self, x_data):
        return np.linalg.norm(self.X_train-x_data, axis = 1)
    
    def calculate_manhattan_distance(self, x_data):
        return np.linalg.norm(self.X_train-x_data, ord = 1, axis = 1)
    
    def calculate_neighbors(self, distances):
        neighbors = []
        k_minimum_elements = np.argpartition(distances,self.k)
#         print(k_minimum_elements)
        for i in range(0, self.k):
            neighbors.append(self.Y_train[k_minimum_elements[i]])
        return neighbors
#         sorted_distance = sorted(zip(distances, self.Y_train))
#         for _, y in sorted_distance:
#             neighbors.append(y)
#         return neighbors
    
    def most_common_neighbors(self, neighbors):
        greater_than_50k = neighbors.count(" >50K")
        less_than_50k = neighbors.count(" <=50K")
        return ">50K" if greater_than_50k > less_than_50k else "<=50K"  
    
    def calculate_positive_rate(self, preficted_values):
        return sum(1 for x in preficted_values if x == ">50K")
    
    def predict(self, test_data):
        prediced_values = []
        for x_data in test_data:
            #1. Calulate the Distances
            if self.dist_cal_method == "manhattan":
                distances = self.calculate_manhattan_distance(x_data)
            else:
                distances = self.calculate_euclidean_distance(x_data)
            
            #2. Calculate the k nearest neighbors compairing distances from the y_train
            neighbors = self.calculate_neighbors(distances)
            
            #3. Return the most common y label of those k nearest neighbors
            prediced_values.append(self.most_common_neighbors(neighbors))
        return prediced_values
    
    def accuracy_score(self, test_data, test_labels):
        prediction = self.predict(test_data)
        
        predictions = [label.strip() for label in prediction]
        test_label =  [label.strip() for label in test_labels]
        
        accuracy_score = sum(p == t for p, t in zip(predictions, test_label) ) / len(test_data)
        return accuracy_score
        
```
